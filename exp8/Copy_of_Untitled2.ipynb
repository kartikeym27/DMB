```python
data= list(df['products'].apply(lambda x:x.split(',')))
print(data)
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

df1 = apriori(df,min_support=0.02,use_colnames=True)
print(df1)
```

```python
df1.sort_values(by="support",ascending = False)

df1["length"] = df1["itemsets"].apply(lambda x:len(x))
print(df1)
```

```python
print(df1[(df1["length"]==2) & (df1["support"]>0.05)])
# print(df1)
```

```python
df1[(df1["length"]==3)& (df1["support"]>0.05)]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

rules = association_rules(df1,metric="confidence",min_threshold=.5)
print(rules)
```
